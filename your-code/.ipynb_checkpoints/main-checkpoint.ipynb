#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users_df= pd.read_csv('users_table.csv')
```

```python
print(users_df.shape)
users_df.head()
```

#### 3. Rename Id column to userId

```python
users_df.rename(columns={'Id':'userId'},inplace=True)
```

```python
users_df.head()
```

#### 4. Import posts table:

```python
table_df= pd.read_csv('posts_table.csv')
```

```python
table_df.head()
```

Rename Id column to postId and OwnerUserId to userId

```python
table_df.rename(columns={'Id':'postId','OwnerUserId':'userId'},inplace=True)
```

```python
print(table_df.shape)
table_df.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
userId = users_df['userId']
Reputation=  users_df['Reputation']
Views= users_df['Views']
UpVotes= users_df['UpVotes']
DownVotes= users_df['DownVotes']
usersdf=pd.concat([userId,Reputation,Views,UpVotes,DownVotes],axis=1)
usersdf.head()
```

```python
postId=table_df['postId']
Score=table_df['Score']
userId=table_df['userId']
ViewCount=table_df['ViewCount']
CommentCount=table_df['CommentCount']
tabledf=pd.concat([postId,Score,userId,ViewCount,CommentCount],axis=1)
tabledf.head()
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
mergeddf=pd.merge(left=usersdf[['userId','Reputation','Views','UpVotes','DownVotes']],
                    right=tabledf[['userId','postId','Score','ViewCount','CommentCount']],
                    left_on ='userId',
                    right_on = 'userId',
                    how='inner'
                    )
```

```python
print(usersdf.shape)
tabledf.shape
```

```python
print(mergeddf.shape)
mergeddf.sample(20)
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
print(mergeddf.shape)
print(mergeddf['userId'].isnull().value_counts())
#as expected, there is no missing values in userId column. let´s check the others
```

```python
print(tabledf['ViewCount'].isnull().value_counts())
print(mergeddf[['Reputation','Views','UpVotes','DownVotes','postId','Score','ViewCount','CommentCount']].isnull().value_counts())
'''There are missing values in viewCount column, which belonged to tabledf'''
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#missing values in ViewCount column are too many to clean. So I should filling them.
print(mergeddf['ViewCount'].isnull().value_counts())
mergeddf['ViewCount'].value_counts()
```

```python
#In this case I am going to fill them with the mean
mergeddf['ViewCount'].fillna(0).value_counts()
```

```python
mergeddf['ViewCount']=mergeddf['ViewCount'].fillna(mergeddf['ViewCount'].mean())
```
